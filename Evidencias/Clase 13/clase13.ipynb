### Simulación en Python

```python
# -------- Imports ------------

from pytc2.cuadripolos import TabcdY_s, TabcdZ_s, Z2Tabcd_s
import sympy as sp
```

```python
# ------------- Desarollo ------------

Za, Zb = sp.symbols('Za, Zb', complex=True)
s = sp.symbols('s', complex=True)

Za = s
Zb = 1/s

T_RG = TabcdZ_s(1)
T_RL = TabcdY_s(1)
T_lattice = Z2Tabcd_s(sp.Matrix([[(Za+Zb)/2, (Zb-Za)/2], [(Zb-Za)/2, (Za+Zb)/2]]))
```

```python
# ------------- Matriz T del circuito ----------------

T_tot = T_RG * T_lattice * T_RL
display( sp.simplify(T_tot) )
```

```python
# ------------- Transferencia ---------------

Vtransf = sp.simplify(sp.expand( 1/T_tot[0,0]))
display(Vtransf)
```

Se puede ver como la simulación se corresponde con los cálculos hechos a mano.
